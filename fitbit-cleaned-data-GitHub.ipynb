### 4.2 Using Google Sheet to Clean Data <a class="anchor" id="using"></a>
Each dataset was cleaned using Google Sheets. The following steps were taken within each dataset: 
1. Sorted and filtered data by Id to obtain how many unique users there were within each dataset. All blank spaces in Id were filtered out
2. Checked for duplicate data using the ‘remove dulicate’ tool in Google Sheets
3. Added a new colum for days of the week in dailyActivity_merged.csv. New column name was labeled as 'activity_day'
4. To find missing information conditional formatting was used. No matching results were found. 
5. Column names for all the files used were then changed to lower case and underscores were inputted to seperate the words. I.e. TotalDistance = total_distance  

     * 1. sleepDay_merged: changed column name 'totaltimeinbed' to 'total_minutes_bed'
     * 2. dailysteps file: changed column name 'steptotal' to 'total_steps' to standardize the column names across tables 
     * 3. hourlysteps table column name 'steptotal' to 'total_steps'     
     
6. Separated Date and Hour into two columns when needed for later analysis. Utilized the 'Text to Columns' tool to do so. The following tables were changed:
   * 1. hourlySteps_merged -- named new columns activity_hour & am/pm
   * 2. hourlyCalories_merged -- named new columns activity_hour & am/pm

After the cleaning process was done, there were only 3 rows in the sleepDay_merged file that had duplicate information.

### 5.10  METs analysis  <a class="anchor" id="METs"></a>

* Q. What is METs?
     * MET: The ratio of the work metabolic rate to the resting metabolic rate. One MET is defined as 1 kcal/kg/hour. The **metabolic equivalent of task (MET)** is the objective measure of the ratio of the rate at which a person expends energy equivalent to the energy cost of sitting down.

Figure 19 [GitHub](https://github.com/alisahial/fitbitcleaned/issues/1)

* MET values here are divided by 10 because according to meta data of the dataset: "All MET values exported from Fitabase are multiplied by 10. Please divide by 10 to get accurate MET values"

```python
--METs

SELECT DISTINCT temp1.id, 
    temp1.activity_minute, 
    SUM(temp1.METs)/24 as avg_mets, 
    temp2.Calories
FROM `fluid-tuner-406721.FitBit_Case_Study.METs_minute` as temp1
INNER JOIN fluid-tuner-406721.FitBit_Case_Study.daily_activity AS temp2
ON temp1.id = temp2.id 
Group By temp1.Id, temp1.activity_minute, temp2.Calories
Order by Id,activity_minute;
```
